```python
#Q1
'''Ordinal encoding assigns unique integers to categories based on their order, while label encoding assigns unique integers without considering any 
order. Use ordinal encoding when the categories have a meaningful sequence (e.g., 'Low', 'Medium', 'High'). Use label encoding when there is no 
intrinsic order among categories (e.g., 'Red', 'Blue', 'Green').'''
```

```python
#Q2
'''Target Guided Ordinal Encoding assigns ordinal values to categories based on the mean of the target variable. Categories with higher means receive 
higher ordinal values. This technique is useful when there's a strong relationship between the categorical feature and the target variable.

Example:
In a project predicting house prices, you have a 'Neighborhood' feature. Calculate the mean house price for each neighborhood and encode them based on 
these means. This ensures neighborhoods with higher average prices get higher ordinal values, leveraging the relationship between neighborhood and 
house price for better predictions.'''
```

```python
#Q4
from sklearn.preprocessing import LabelEncoder
data = {
    'Color': ['red', 'green', 'blue'],
    'Size': ['small', 'medium', 'large'],
    'Material': ['wood', 'metal', 'plastic']
}

encoder = LabelEncoder()

data["encoded_color"] = encoder.fit_transform(data["Color"] )
data["encoded_Size"] = encoder.fit_transform(data["Size"] )
data["encoded_Material"] = encoder.fit_transform(data["Material"] )

print(data)
```

```python
#Q5
import numpy as np
import pandas as pd

data = {
    'Age': [25, 32, 47, 51, 62],
    'Income': [50000, 60000, 80000, 85000, 90000],
    'Education_Level': [12, 14, 16, 16, 18]
}
df = pd.DataFrame(data)

cov_matrix = np.cov(df.T)

print(cov_matrix)
```

```python
#Q7
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
data = {
    'Temperature': [22, 24, 19, 23, 25, 21],
    'Humidity': [30, 45, 55, 60, 35, 50],
    'Weather Condition': ['Sunny', 'Cloudy', 'Rainy', 'Sunny', 'Cloudy', 'Rainy'],
    'Wind Direction': ['North', 'South', 'East', 'West', 'North', 'East']
}

df = pd.DataFrame(data)
OHE = OneHotEncoder()
weencoded = OHE.fit_transform(df[['Weather Condition']])
wetherdf= pd.DataFrame(weencoded.toarray(), columns=OHE.get_feature_names_out(['Weather Condition'] ) )
wiencoded = OHE.fit_transform(df[['Wind Direction']])
witherdf= pd.DataFrame(wiencoded.toarray(), columns=OHE.get_feature_names_out(['Wind Direction'] ) )
con_df = pd.concat([wetherdf,witherdf],axis=False )
con_df
final_df = pd.concat([df,con_df],axis=False).drop(columns=[ 'Weather Condition','Wind Direction'])
final_df
```
